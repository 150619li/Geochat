```mermaid
sequenceDiagram
    participant 用户
    participant 前端
    participant API
    participant 服务层
    participant 数据库

    %% 注册流程
    用户->>前端: 输入注册信息
    前端->>API: POST /api/users/register
    API->>服务层: 创建用户
    服务层->>数据库: 保存用户数据
    数据库-->>服务层: 确认保存
    服务层-->>API: 返回结果
    API-->>前端: 返回注册状态
    前端-->>用户: 显示注册结果

    %% 登录流程
    用户->>前端: 输入登录信息
    前端->>API: POST /api/users/login
    API->>服务层: 验证用户
    服务层->>数据库: 查询用户信息
    数据库-->>服务层: 返回用户数据
    服务层->>服务层: 验证密码
    服务层->>服务层: 生成JWT Token
    服务层-->>API: 返回Token
    API-->>前端: 返回Token和用户信息
    前端->>前端: 存储Token
    前端-->>用户: 跳转主页

```

```mermaid
sequenceDiagram
    participant 用户
    participant 前端
    participant API
    participant 服务层
    participant 数据库

    %% 群聊消息发送流程
    用户->>前端: 发送消息
    前端->>API: POST /api/messages/send
    API->>服务层: 处理消息
    服务层->>数据库: 保存消息
    数据库-->>服务层: 确认保存
    服务层-->>API: 返回结果
    API-->>前端: 推送消息
    前端-->>用户: 显示消息

    %% 群聊消息展示流程
    用户->>前端: 进入群聊
    前端->>API: GET /api/messages/groupId
    API->>服务层: 获取群消息
    服务层->>数据库: 查询消息历史
    数据库-->>服务层: 返回消息列表
    服务层-->>API: 返回消息数据
    API-->>前端: 返回消息列表
    前端->>前端: 格式化消息
    前端-->>用户: 展示消息历史

```

```mermaid
sequenceDiagram
    participant 用户
    participant 前端
    participant API
    participant 服务层
    participant 数据库

    %% 修改地理位置流程
    用户->>前端: 更新位置信息
    前端->>前端: 获取GPS坐标
    前端->>API: PUT /api/users/location
    API->>服务层: 更新位置
    服务层->>数据库: 保存新位置
    数据库-->>服务层: 确认更新
    服务层->>服务层: 检查nearby群组
    服务层-->>API: 返回更新结果
    API-->>前端: 返回状态和nearby群组
    前端-->>用户: 显示位置更新状态
```
